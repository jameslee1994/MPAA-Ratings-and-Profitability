```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/ajaverett/data/refs/heads/main/titanic.csv')
```

```python
def titanic_age_now(age=12):
  '''Calculates how old a person on the titanic would be now.'''
  current_year = 2025
  titanic_sunk = 1912
  birth_year = titanic_sunk - age
  person_age = current_year - birth_year
  return person_age

print(titanic_age_now(58))
```

```python
df["Age_now"] = df["Age"].apply(titanic_age_now)
df.sort_values(by="Age_now", ascending=False).head(20)
```
